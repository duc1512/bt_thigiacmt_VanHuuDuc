```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ng dn nh
my_path='C:\\Users\\Vu Duc Admin\\Documents\\img\\02.jpg'
img = Image.open(my_path)
plt.imshow(img)
```

```python
# Chuyn sang nh xm
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
